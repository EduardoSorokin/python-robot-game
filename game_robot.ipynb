### Legend

- **R:** Robot (You)
- **C:** Coin (Reward)
- **G:** Gas (Reward)
- **W:** Weapon (Reward)

```python
class Point():
    
    def __init__(self, x, y):
        self.x = x
        self.y = y
        
    def __str__(self):
        return '<%s, %s>' % (self.x, self.y)


class Reward(Point):
    
    def __init__(self, x, y, name):
        super(Reward, self).__init__(x, y)
        self.name = name

    def __str__(self):
        return '<%s, %s>: %s' % (self.x, self.y, self.name)
    
    def __repr__(self):
        return '<Reward> %s' % str(self)
        

class Robot(Point):
        
    def move_up(self):
        if self.y > 0:
            self.y -= 1
            return True
        else:
            print("Movement not allowed.")
            return False
        
    def move_down(self):
        if self.y < 10:
            self.y += 1
            return True
        else:
            print("Movement not allowed.")
            return False
    
    def move_left(self):
        if self.x > 0:
            self.x -= 1
            return True
        else:
            print("Movement not allowed.")
            return False
    
    def move_right(self):
        if self.x < 10:
            self.x += 1
            return True
        else:
            print("Movement not allowed.")
            return False

# -------------------------------------------------------------------

def check_reward(robot, rewards):
    found = False
    
    for reward in rewards:
        if reward.x == robot.x and reward.y == robot.y:
            print('The robot found the reward: "%s"\n' % reward.name)
            found = True
            
    return found

# -------------------------------------------------------------------

import random
r1 = Reward(random.randint(0, 10), random.randint(0, 10), 'coin')
r2 = Reward(random.randint(0, 10), random.randint(0, 10), 'gas')
r3 = Reward(random.randint(0, 10), random.randint(0, 10), 'weapon')
r4 = Reward(random.randint(0, 10), random.randint(0, 10), 'coin')
r5 = Reward(random.randint(0, 10), random.randint(0, 10), 'gas')
r6 = Reward(random.randint(0, 10), random.randint(0, 10), 'weapon')
r7 = Reward(random.randint(0, 10), random.randint(0, 10), 'coin')
r8 = Reward(random.randint(0, 10), random.randint(0, 10), 'gas')
r9 = Reward(random.randint(0, 10), random.randint(0, 10), 'weapon')
rewards = [r1, r2, r3, r4, r5, r6, r7, r8, r9]

# --------------------------------------------------------------------

robot = Robot(random.randint(0, 10), random.randint(0, 10))
print("\nThe robot started at: %s\n" % robot)
check_reward(robot, rewards)

# --------------------------------------------------------------------

i = 0
while i < 10:
    moviment = input("Type 'up', 'down', 'left' or 'right' to move: ")
    
    if moviment == 'up':
        if not robot.move_up():
            continue
    elif moviment == 'down':
        if not robot.move_down():
            continue
    elif moviment == 'left':
        if not robot.move_left():
            continue
    elif moviment == 'right':
        if not robot.move_right():
            continue
    else:
        print('Invalid moviment!\n')
        continue
    
    print("The robot is now at: %s\n" % robot)
    check_reward(robot, rewards)
    i += 1

print("\nNo more moves!")
print("\nTHE END")
```
